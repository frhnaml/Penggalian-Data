```python
import pandas as pd


csv_file = 'International Logistic Association Memberships.xlsx - Association ABC memberships.csv'

df_csv = pd.read_csv(csv_file)

print("Nama kolom dalam dataset:")
print(df_csv.columns)

# 2. Menampilkan 7 baris data awal dan informasi dasar dataset
print("\nData dari file CSV:")
print(df_csv.head(7))
print(df_csv.info())

# 3. Menghitung jumlah dan rata-rata iuran
# Menghitung jumlah dan rata-rata iuran jika kolom 'Dues amount' ada
if 'Dues amount' in df_csv.columns:
    df_csv['Dues amount'] = df_csv['Dues amount'].replace(r'[\$,]', '', regex=True).astype(float)
    total_dues = df_csv['Dues amount'].sum()
    avg_dues = df_csv['Dues amount'].mean()
    print(f"\nTotal iuran dari file CSV: {total_dues}")
    print(f"Rata-rata iuran dari file CSV: {avg_dues}")
else:
    print("\nKolom 'Dues amount' tidak ditemukan dalam dataset. Periksa nama kolom yang benar.")
```
